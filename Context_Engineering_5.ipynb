Load and chunk your document

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter

loader = PyPDFLoader("/india_ai_strategy.pdf")
docs = loader.load()

splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=100
)

chunks = splitter.split_documents(docs)

print(f"✅ Loaded {len(chunks)} chunks")
```

step 3

```python
# Install the free embedding model
!pip install sentence-transformers

from langchain.embeddings import HuggingFaceEmbeddings

# Use a free Hugging Face model
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

vectorstore = FAISS.from_documents(chunks, embeddings)

print("✅ Vector store created with Hugging Face embeddings (no OpenAI key needed)")
```

step 4 select context

```python
# ✅ Step 4: Selecting context with Hugging Face embeddings + FAISS

# ----------------------------------------------
# Recap: In Step 3 you did this:
# ----------------------------------------------
# from langchain.embeddings import HuggingFaceEmbeddings
# embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
# vectorstore = FAISS.from_documents(chunks, embeddings)
# ----------------------------------------------

# ✅ 1️⃣ Create a retriever from the vectorstore
retriever = vectorstore.as_retriever(
    search_kwargs={"k": 5}  # You can change 'k' to get more or fewer chunks
)

# ✅ 2️⃣ Define your question
query = "What are the key priorities of India's AI Strategy?"

# ✅ 3️⃣ Retrieve relevant chunks by semantic similarity
relevant_docs = retriever.get_relevant_documents(query)

# ✅ 4️⃣ Check what you got
print(f"✅ Retrieved {len(relevant_docs)} relevant chunks\n")

for i, doc in enumerate(relevant_docs):
    print(f"--- Chunk #{i+1} ---")
    print(doc.page_content[:500])  # Print first 500 chars to preview
    print("\n")
```

step 5 Compresss context

```python
# ✅ STEP 5: Compress context using Hugging Face summarization

# ----------------------------------------------
# Install transformers if you haven’t
# ----------------------------------------------
!pip install transformers sentencepiece

from transformers import pipeline

# ✅ 1️⃣ Load a summarization pipeline
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

# ✅ 2️⃣ Combine retrieved chunks into one big text
all_text = "\n\n".join([doc.page_content for doc in relevant_docs])

# ✅ 3️⃣ Optionally check length
print(f"✅ Total input chars: {len(all_text)}")

# ✅ 4️⃣ Run summarization
# Note: Transformers pipelines often have a max input length (~1024 tokens for BART)
# So we may need to chunk if text is too long

# If too long, take the first ~3000 chars as a simple example:
input_text = all_text[:3000]

summary = summarizer(
    input_text,
    max_length=200,   # Max output tokens
    min_length=50,    # Min output tokens
    do_sample=False
)

# ✅ 5️⃣ Print the summary
print("\n✅ Compressed Summary:\n")
print(summary[0]['summary_text'])
```

Step 6 : Isolating Context

```python
# ✅ STEP 6: Isolate context — extract keywords + rewrite final answer

from transformers import pipeline

# ---------------------------------------------------
# 1️⃣ Use a simple Hugging Face LLM for keyword extraction
# ---------------------------------------------------

# We’ll reuse the summarizer or you can load a small LLM
# For simplicity, we’ll use a simple fill-mask or text generation pipeline

# But here’s a trick: we can prompt a summarizer for keywords too.
# For a more advanced approach you’d use a local LLM like Llama 3.

# ---------------------------------------------------
# 2️⃣ Build a simple prompt for keywords
# ---------------------------------------------------

# Use the same summarizer pipeline for short text generation
# (Not perfect but works for simple keyword extraction)

prompt_for_keywords = f"""
Text: {summary[0]['summary_text']}

Extract 5 important keywords from the above text, separated by commas:
"""

# Use text generation pipeline
from transformers import AutoModelForCausalLM, AutoTokenizer

# We'll use a small open-source model for short generation (e.g., GPT-2)
model_name = "gpt2"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)

from transformers import pipeline

generator = pipeline("text-generation", model=model, tokenizer=tokenizer)

# Generate keywords
keyword_output = generator(
    prompt_for_keywords,
    max_length=50,
    num_return_sequences=1
)

keywords = keyword_output[0]['generated_text'].split("Extract 5 important keywords from the above text, separated by commas:")[-1].strip()

# ---------------------------------------------------
# 3️⃣ Optionally rewrite the summary to final answer
# ---------------------------------------------------

rewrite_prompt = f"""
Rewrite this text in under 100 words:

{summary[0]['summary_text']}
"""

rewrite_output = generator(
    rewrite_prompt,
    max_length=150,
    num_return_sequences=1
)

final_answer = rewrite_output[0]['generated_text'].split("Rewrite this text in under 100 words:")[-1].strip()

# ✅ ✅ ✅ DONE — print it all!
print("\n✅ Isolated Context:")
print(f"🔑 Keywords: {keywords}")
print(f"\n📝 Final Answer: {final_answer}")
```
